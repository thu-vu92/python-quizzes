## Basic Data Types
1. Integer: A whole number without a fractional part. E.g., 100, 156, 2012412
2. Float: A number with a fractional part. E.g., 0.75, 3.1415, 961.1241250, 7/8
3. Boolean: A binary value. It can be either True or False.
4. String: A sequence of Unicode characters (e.g., 'hello', 'R2D2', 'Tomi', '124.56.128.41')

### Try/Catch Error Handling

Handling exceptions to avoid crashes.

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
# Multiple Exceptions
try:
    x = int(input("Enter a number: "))
    y = 10 / x
except ValueError as e:
    print(f"Value Error: {e} - Please enter a valid integer.")
except ZeroDivisionError as e:
    print(f"Zero Division Error: {e} - Division by zero is not allowed.")
```

```python
#  Multiple Exceptions with Else and Finally
try:
    num = int(input("Enter an integer: "))
    print(f"100 / {num} = {100 / num}")
except ValueError as e:
    print(f"Value Error: {e} - Please enter a valid integer.")
except ZeroDivisionError as e:
    print(f"Zero Division Error: {e} - Division by zero is not allowed.")
else:
    print("The operation was successful.")
finally:
    print("This block always executes, regardless of an exception.")
```

```python
try:
    filename = input("Enter a file name: ")
    with open(filename, 'r') as file:
        data = file.read()
    number = int(data)
    result = 100 / number
except FileNotFoundError as e:
    print(f"File Not Found Error: {e} - The specified file does not exist.")
except ValueError as e:
    print(f"Value Error: {e} - The file does not contain a valid integer.")
except ZeroDivisionError as e:
    print(f"Zero Division Error: {e} - Division by zero is not allowed.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

### Zip
Combines two or more iterables element-wise.

```python
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]
combined = list(zip(names, ages))
print(combined)
```

```python
# Unzipping
combined = [('Alice', 25), ('Bob', 30), ('Charlie', 35)]
names, ages = zip(*combined)
```

```python
print(names, ages)
```

```python
# Zip dictionaries
dict_a = {'a': 1, 'b': 2, 'c': 3}
dict_b = {'x': 4, 'y': 5, 'z': 6}

keys = list(dict_a.keys())
values = list(dict_b.values())

combined = list(zip(keys, values))
print(combined)
```

```python
# Create Dictionary
keys = ['name', 'age', 'city']
values = ['Alice', 30, 'Toronto']

combined_dict = dict(zip(keys, values))
print(combined_dict)
```

```python
# Zip in a loop
names = ['Thu', 'Fidel', 'Charlie']
scores = [100, 90, 95]

for name, score in zip(names, scores):
    print(f"{name} scored {score}")
```

```python
# Zip with enumearate
names = ['Thy', 'Fidel', 'Charlie']
scores = [100, 90, 95]

for i, (name, score) in enumerate(zip(names, scores)):
    print(f"{i + 1}. {name} scored {score}")
```
